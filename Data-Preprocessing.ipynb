```python
#importing libraries
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Data.csv")
```

```python
# glucose column contains value from o-69  replace it to mean
# BloodPressure columns contain value from 0-79  replace it to mean
#SkinThickness :- remove values below 9  replace it to mean
#Insulin:-remove values below 16  replace it to mean
#BMI:-remove values below 18  replace it to mean

# remaining columns are good and no need to update 

```

```python
# Replace all Glucose values < 70 with NaN
df.loc[df['Glucose'] < 70, 'Glucose'] = np.nan
# Then fill NaN with the mean of valid values
df['Glucose'] = df['Glucose'].fillna(df['Glucose'].mean())
```

```python
# Replace all BloodPressure values < 70 with NaN
df.loc[df['BloodPressure'] < 80, 'BloodPressure'] = np.nan
# Then fill NaN with the mean of valid values
df['BloodPressure'] = df['BloodPressure'].fillna(df['BloodPressure'].mean())
```

```python
# Replace all SkinThickness values < 70 with NaN
df.loc[df['SkinThickness'] < 10, 'SkinThickness'] = np.nan
# Then fill NaN with the mean of valid values
df['SkinThickness'] = df['SkinThickness'].fillna(df['SkinThickness'].mean())
```

```python
# Replace all Insulin values < 70 with NaN
df.loc[df['Insulin'] < 17, 'Insulin'] = np.nan
# Then fill NaN with the mean of valid values
df['Insulin'] = df['Insulin'].fillna(df['Insulin'].mean())
```

```python
# Replace all BMI values < 70 with NaN
df.loc[df['BMI'] < 17, 'BMI'] = np.nan
# Then fill NaN with the mean of valid values
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())
```

```python
df = df.reset_index(drop=True)
```

```python
cols_to_round = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI']  # Example list
df[cols_to_round] = df[cols_to_round].round(1)
```
