```python
# Set modern design
sns.set_style("whitegrid")

# Count values
outcome_counts = df["Outcome"].value_counts().sort_index()

# Create figure
plt.figure(figsize=(8, 5))
bars = plt.bar(outcome_counts.index.astype(str), outcome_counts.values, color='mediumseagreen', edgecolor='black')

# Add labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 2, f'{int(height)}', 
             ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

# Add titles and labels
plt.title("Diabetes Cases by Outcome", fontsize=16, fontweight='bold')
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)", fontsize=13)
plt.ylabel("Count", fontsize=13)
plt.xticks(rotation=0)

# Final layout
plt.tight_layout()
plt.show()
```

```python

# Set modern style
sns.set_style("whitegrid")

# Create glucose bins (you can adjust ranges as needed)
bins = [0, 99, 125, 140, 200]
labels = ['Low (<100)', 'Pre-Diabetic (100-125)', 'Borderline (126-140)', 'High (>140)']
df['GlucoseRange'] = pd.cut(df['Glucose'], bins=bins, labels=labels, include_lowest=True)

# Count values in each bin
glucose_counts = df['GlucoseRange'].value_counts().sort_index()

# Create pie chart
plt.figure(figsize=(8, 8))
colors = sns.color_palette("pastel")
plt.pie(glucose_counts, labels=glucose_counts.index, autopct='%1.1f%%',
        startangle=140, colors=colors, wedgeprops={'edgecolor': 'black'})

# Title
plt.title("Glucose Level Distribution", fontsize=16, fontweight='bold')

# Show chart
plt.tight_layout()
plt.show()
```

```python
# Set modern style
sns.set_style("whitegrid")

# Create bins and labels for Pregnancies
bins = [-1, 0, 2, 5, df['Pregnancies'].max()]  # -1 to include 0
labels = ['None', '1–2', '3–5', '6+']
df['PregnancyRange'] = pd.cut(df['Pregnancies'], bins=bins, labels=labels)

# Count values in each bin
preg_counts = df['PregnancyRange'].value_counts().sort_index()

# Create pie chart
plt.figure(figsize=(8, 8))
colors = sns.color_palette("pastel")
plt.pie(preg_counts, labels=preg_counts.index, autopct='%1.1f%%',
        startangle=140, colors=colors, wedgeprops={'edgecolor': 'black'})

# Title
plt.title("Distribution of Pregnancies", fontsize=16, fontweight='bold')

# Show the chart
plt.tight_layout()
plt.show()
```

```python


# Set modern style
sns.set_style("whitegrid")

# Count number of records for each pregnancy value
preg_counts = df['Pregnancies'].value_counts().sort_index()

# Create figure
plt.figure(figsize=(10, 6))
bars = plt.bar(preg_counts.index.astype(str), preg_counts.values,
               color='skyblue', edgecolor='black')

# Add labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{int(height)}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Add titles and labels
plt.title("Number of Records by Pregnancy Count", fontsize=16, fontweight='bold')
plt.xlabel("Pregnancies", fontsize=13)
plt.ylabel("Number of Patients", fontsize=13)
plt.xticks(rotation=0)

# Final layout
plt.tight_layout()
plt.show()
```

```python
# Set modern style
sns.set_style("whitegrid")

# Create figure with 3 subplots for different columns
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=False)

# Swarmplot for BMI vs Outcome
sns.swarmplot(ax=axes[0], data=df, x='Outcome', y='BMI', palette='Set2')
axes[0].set_title("BMI by Diabetes Outcome", fontsize=14, fontweight='bold')
axes[0].set_xlabel("Outcome (0 = No, 1 = Yes)")
axes[0].set_ylabel("BMI")

# Swarmplot for Age vs Outcome
sns.swarmplot(ax=axes[1], data=df, x='Outcome', y='Age', palette='Set1')
axes[1].set_title("Age by Diabetes Outcome", fontsize=14, fontweight='bold')
axes[1].set_xlabel("Outcome")
axes[1].set_ylabel("Age")

# Swarmplot for Glucose vs Outcome
sns.swarmplot(ax=axes[2], data=df, x='Outcome', y='Glucose', palette='Set3')
axes[2].set_title("Glucose by Diabetes Outcome", fontsize=14, fontweight='bold')
axes[2].set_xlabel("Outcome")
axes[2].set_ylabel("Glucose Level")

# Final layout
plt.tight_layout()
plt.show()
```
